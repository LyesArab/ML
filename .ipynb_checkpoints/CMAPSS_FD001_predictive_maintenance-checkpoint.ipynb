# Predictive Maintenance — NASA C-MAPSS FD001 (Remaining Useful Life)

**Objectif** : prédire le nombre de cycles restants avant panne (RUL — Remaining Useful Life)
pour des moteurs turbofan, à partir de mesures capteurs multivariées.

**Dataset** : NASA C-MAPSS, sous-ensemble **FD001**
- 100 trajectoires d'entraînement (run-to-failure complet)
- 100 trajectoires de test (arrêtées avant la panne)
- 1 seule condition opérationnelle (niveau de la mer), 1 seul mode de panne (dégradation HPC)
- Vraie vérité terrain fournie (`RUL_FD001.txt`) — contrairement à un problème de détection
  d'anomalie non supervisé, ici on peut calculer de vraies métriques de régression.

**Structure du notebook** :
1. Chargement & EDA
2. Feature engineering (RUL target, rolling stats, sélection de capteurs)
3. Baseline — modèle tabulaire (Random Forest / XGBoost) sur le dernier cycle de chaque unité
4. Modèle séquentiel — LSTM sur fenêtres glissantes
5. Évaluation (RMSE + score PHM08 officiel, asymétrique)
6. Visualisations & conclusion


## Phase 1 — Setup & Chargement des données

⚠️ Place `train_FD001.txt`, `test_FD001.txt` et `RUL_FD001.txt` dans le même dossier
que ce notebook (ou modifie `DATA_DIR` ci-dessous).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = "."  # <-- change si besoin

index_names = ['unit', 'cycle']
setting_names = ['os1', 'os2', 'os3']
sensor_names = [f'sensor_{i}' for i in range(1, 22)]
col_names = index_names + setting_names + sensor_names

train = pd.read_csv(f"{DATA_DIR}/train_FD001.txt", sep=r"\s+", header=None, names=col_names)
test = pd.read_csv(f"{DATA_DIR}/test_FD001.txt", sep=r"\s+", header=None, names=col_names)
rul_test = pd.read_csv(f"{DATA_DIR}/RUL_FD001.txt", sep=r"\s+", header=None, names=["RUL"])

print("Train shape:", train.shape)
print("Test shape :", test.shape)
print("RUL test shape:", rul_test.shape)
print("Nombre d'unités (moteurs) train:", train['unit'].nunique())
print("Nombre d'unités (moteurs) test :", test['unit'].nunique())
train.head()
```

### 1.1 — Exploration rapide

Combien de cycles vit chaque moteur avant panne (dans le train, on a le run-to-failure complet) ?

```python
life = train.groupby("unit")["cycle"].max()
print(life.describe())

plt.figure(figsize=(8,4))
plt.hist(life, bins=20, edgecolor="black")
plt.xlabel("Durée de vie (cycles)")
plt.ylabel("Nombre de moteurs")
plt.title("Distribution des durées de vie — train FD001")
plt.show()
```

Vérifions les capteurs à variance quasi nulle (comme `throttle_pos` sur ton dataset OBD —
même logique ici : un capteur constant n'apporte aucune info et peut fausser un scaler).

```python
sensor_stats = train[sensor_names].agg(['mean', 'std'])
print(sensor_stats.T)

low_var_sensors = sensor_stats.loc['std'][sensor_stats.loc['std'] < 1e-4].index.tolist()
print("\nCapteurs à variance quasi nulle (à exclure):", low_var_sensors)
```

```python
# Visualisation de quelques capteurs pour un moteur donné (dégradation visible en fin de vie)
unit_example = 1
df_u = train[train["unit"] == unit_example]

fig, axes = plt.subplots(2, 2, figsize=(12, 7))
for ax, s in zip(axes.flat, ["sensor_2", "sensor_3", "sensor_4", "sensor_7"]):
    if s in df_u.columns:
        ax.plot(df_u["cycle"], df_u[s])
        ax.set_title(s)
        ax.set_xlabel("cycle")
plt.suptitle(f"Évolution des capteurs — moteur {unit_example}")
plt.tight_layout()
plt.show()
```

## Phase 2 — Construction de la target RUL & feature engineering

**RUL (train)** = `cycle_max_du_moteur - cycle_courant` (à la fin de vie, RUL=0).

**Clipping** : c'est une pratique standard sur C-MAPSS. Au début de vie, le RUL "vrai" est
énorme et sans rapport avec l'état de dégradation réel (aucun capteur ne peut distinguer
RUL=250 de RUL=280 — le moteur est juste en bon état). On plafonne donc le RUL à une valeur
raisonnable (125 est la valeur la plus utilisée dans la littérature FD001) pour ne pas que le
modèle essaie d'apprendre l'impossible sur la portion "normale" de la vie.

```python
RUL_CLIP = 125

max_cycle = train.groupby("unit")["cycle"].transform("max")
train["RUL"] = max_cycle - train["cycle"]
train["RUL_clipped"] = train["RUL"].clip(upper=RUL_CLIP)

train[["unit", "cycle", "RUL", "RUL_clipped"]].head(10)
```

```python
feature_sensors = [s for s in sensor_names if s not in low_var_sensors]
print(f"Capteurs retenus ({len(feature_sensors)}):", feature_sensors)

def add_rolling(df, cols, window=5):
    df = df.copy()
    for c in cols:
        df[f"{c}_roll_mean"] = df.groupby("unit")[c].transform(
            lambda x: x.rolling(window, min_periods=1).mean())
        df[f"{c}_roll_std"] = df.groupby("unit")[c].transform(
            lambda x: x.rolling(window, min_periods=1).std().fillna(0))
    return df

train_fe = add_rolling(train, feature_sensors)
test_fe = add_rolling(test, feature_sensors)

feature_cols = (feature_sensors + setting_names
                + [f"{c}_roll_mean" for c in feature_sensors]
                + [f"{c}_roll_std" for c in feature_sensors])
print(f"\nNombre total de features: {len(feature_cols)}")
```

## Phase 3 — Baseline : modèle tabulaire (Random Forest)

**Protocole standard C-MAPSS pour le test** : pour chaque moteur du test, on ne dispose que
d'une trajectoire tronquée — le vrai RUL fourni correspond au **dernier cycle observé**. On
entraîne donc sur toutes les lignes du train (chaque cycle a son propre RUL), mais on évalue
uniquement sur le dernier cycle de chaque unité de test (c'est la convention du challenge PHM08).

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(train_fe[feature_cols])
y_train = train_fe["RUL_clipped"].values

test_last = test_fe.groupby("unit").last().reset_index()
X_test = scaler.transform(test_last[feature_cols])
y_test = rul_test["RUL"].values

print("X_train:", X_train.shape, " X_test:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

rf = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, pred_rf))
mae_rf = mean_absolute_error(y_test, pred_rf)

print(f"Random Forest — RMSE: {rmse_rf:.2f}   MAE: {mae_rf:.2f}")
```

### (Optionnel) XGBoost — souvent légèrement meilleur que RF sur ce type de données tabulaires

```python
try:
    from xgboost import XGBRegressor
    xgb = XGBRegressor(n_estimators=300, max_depth=6, learning_rate=0.05,
                        subsample=0.8, colsample_bytree=0.8, random_state=42)
    xgb.fit(X_train, y_train)
    pred_xgb = xgb.predict(X_test)
    rmse_xgb = np.sqrt(mean_squared_error(y_test, pred_xgb))
    mae_xgb = mean_absolute_error(y_test, pred_xgb)
    print(f"XGBoost — RMSE: {rmse_xgb:.2f}   MAE: {mae_xgb:.2f}")
except ImportError:
    print("xgboost non installé — `pip install xgboost` pour l'activer")
```

## Phase 4 — Le score PHM08 (métrique officielle du challenge)

Le RMSE traite sur-estimation et sous-estimation du RUL de façon symétrique. En maintenance
prédictive réelle, **sous-estimer le RUL** (dire "il va casser bientôt" alors qu'il a encore du
temps) coûte une intervention de maintenance inutile — ennuyeux mais pas dangereux. **Sur-estimer**
le RUL (dire "il tiendra encore longtemps" alors qu'il va casser) peut mener à une panne en
service — bien plus coûteux/dangereux. Le score PHM08 pénalise donc plus fortement la
sur-estimation (asymétrie exponentielle).

```python
def phm_score(y_true, y_pred):
    """Score officiel PHM08. Plus bas = meilleur. Asymétrique :
    pénalise plus fort la sur-estimation du RUL (y_pred > y_true)."""
    d = y_pred - y_true
    s = np.where(d < 0, np.exp(-d/13) - 1, np.exp(d/10) - 1)
    return np.sum(s)

print(f"Random Forest — PHM score: {phm_score(y_test, pred_rf):.1f}")
try:
    print(f"XGBoost       — PHM score: {phm_score(y_test, pred_xgb):.1f}")
except NameError:
    pass
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, pred_rf, alpha=0.7, label="Random Forest")
lims = [0, max(y_test.max(), pred_rf.max()) + 10]
plt.plot(lims, lims, 'k--', alpha=0.5, label="Prédiction parfaite")
plt.xlabel("RUL réel")
plt.ylabel("RUL prédit")
plt.title("RUL prédit vs réel — Random Forest (test FD001)")
plt.legend()
plt.show()
```

## Phase 5 — Modèle séquentiel (LSTM)

Contrairement au Random Forest (qui ne voit qu'un seul cycle à la fois), le LSTM reçoit une
**fenêtre glissante** de `SEQ_LEN` cycles consécutifs — il peut donc apprendre la *dynamique*
de dégradation, pas juste un état instantané.

```python
SEQ_LEN = 30  # nombre de cycles d'historique par séquence

def gen_sequences(df, feature_cols, seq_len, target_col="RUL_clipped"):
    seqs, targets = [], []
    for uid, g in df.groupby("unit"):
        data = g[feature_cols].values
        rul = g[target_col].values
        if len(data) < seq_len:
            continue  # moteur trop court pour former une séquence complète
        for i in range(len(data) - seq_len + 1):
            seqs.append(data[i:i + seq_len])
            targets.append(rul[i + seq_len - 1])
    return np.array(seqs), np.array(targets)

train_fe_scaled = train_fe.copy()
train_fe_scaled[feature_cols] = scaler.transform(train_fe[feature_cols])

X_seq, y_seq = gen_sequences(train_fe_scaled, feature_cols, SEQ_LEN)
print("X_seq:", X_seq.shape, " y_seq:", y_seq.shape)

from sklearn.model_selection import train_test_split
X_seq_train, X_seq_val, y_seq_train, y_seq_val = train_test_split(
    X_seq, y_seq, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping

lstm_model = models.Sequential([
    layers.Input(shape=(SEQ_LEN, len(feature_cols))),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.2),
    layers.LSTM(32),
    layers.Dropout(0.2),
    layers.Dense(16, activation="relu"),
    layers.Dense(1)
])
lstm_model.compile(optimizer="adam", loss="mse")
lstm_model.summary()
```

```python
early_stop = EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True)

history = lstm_model.fit(
    X_seq_train, y_seq_train,
    validation_data=(X_seq_val, y_seq_val),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="val loss")
plt.xlabel("Epoch"); plt.ylabel("MSE"); plt.legend()
plt.title("LSTM — Courbe d'apprentissage")
plt.show()
```

### Évaluation du LSTM sur le test

Même protocole que pour le RF : on prend, pour chaque moteur de test, sa **dernière fenêtre**
de `SEQ_LEN` cycles (les moteurs trop courts sont ignorés — signale-le si ça arrive souvent,
ça indiquerait qu'il faut réduire `SEQ_LEN`).

```python
def last_sequence_per_unit(df, feature_cols, seq_len):
    seqs, units = [], []
    for uid, g in df.groupby("unit"):
        data = g[feature_cols].values
        if len(data) < seq_len:
            units.append(uid)  # trop court -> exclu, à surveiller
            continue
        seqs.append(data[-seq_len:])
    return np.array(seqs), units

test_fe_scaled = test_fe.copy()
test_fe_scaled[feature_cols] = scaler.transform(test_fe[feature_cols])

X_test_seq, excluded_units = last_sequence_per_unit(test_fe_scaled, feature_cols, SEQ_LEN)
print(f"Séquences de test valides: {X_test_seq.shape[0]} / {test_fe['unit'].nunique()}")
if excluded_units:
    print(f"⚠️ Unités exclues (trop courtes pour SEQ_LEN={SEQ_LEN}): {excluded_units}")

pred_lstm = lstm_model.predict(X_test_seq).flatten()

# aligner y_test avec les unités effectivement gardées (certaines peuvent être exclues
# si trop courtes pour SEQ_LEN)
all_units = sorted(test_fe['unit'].unique())
kept_units = [u for u in all_units if u not in excluded_units]
y_test_lstm = rul_test["RUL"].values[[all_units.index(u) for u in kept_units]]

rmse_lstm = np.sqrt(mean_squared_error(y_test_lstm, pred_lstm))
mae_lstm = mean_absolute_error(y_test_lstm, pred_lstm)
print(f"LSTM — RMSE: {rmse_lstm:.2f}   MAE: {mae_lstm:.2f}")
print(f"LSTM — PHM score: {phm_score(y_test_lstm, pred_lstm):.1f}")
```

## Phase 6 — Comparaison des modèles & conclusion

```python
results = pd.DataFrame({
    "Modèle": ["Random Forest", "LSTM"],
    "RMSE": [rmse_rf, rmse_lstm],
    "PHM score": [phm_score(y_test, pred_rf), phm_score(y_test_lstm, pred_lstm)]
})
try:
    results.loc[len(results)] = ["XGBoost", rmse_xgb, phm_score(y_test, pred_xgb)]
except NameError:
    pass

print(results.sort_values("RMSE"))
```

**Points à commenter une fois exécuté sur les vraies données** :
- Quel modèle a le meilleur RMSE ? Le meilleur PHM score (souvent différent — un modèle peut
  avoir un RMSE correct mais sur-estimer systématiquement, ce qui est puni plus fort par PHM08) ?
- Le LSTM apporte-t-il un vrai gain vs le Random Forest (qui n'a accès qu'à 1 cycle + ses
  rolling stats) ? Si non, ça vaut le coup de le dire honnêtement — un modèle plus simple qui
  fait aussi bien est un résultat valide, pas un échec.
- Regarde les moteurs avec la plus grosse erreur individuelle (`np.abs(pred - y_test)`) : y a-t-il
  un pattern (moteurs avec une dégradation atypique, ou juste peu de cycles observés en test) ?
- **Prochaines étapes possibles** : tester différentes valeurs de `SEQ_LEN` et `RUL_CLIP`,
  essayer un modèle Transformer/attention, ou étendre à FD002/FD004 (multi-conditions) une
  fois FD001 maîtrisé.

```python
err_rf = pred_rf - y_test
err_lstm = pred_lstm - y_test_lstm

plt.figure(figsize=(9, 5))
plt.hist(err_rf, bins=20, alpha=0.5, label=f"Random Forest (RMSE={rmse_rf:.1f})", color="tab:blue")
try:
    err_xgb = pred_xgb - y_test
    plt.hist(err_xgb, bins=20, alpha=0.5, label=f"XGBoost (RMSE={rmse_xgb:.1f})", color="tab:orange")
except NameError:
    pass
plt.hist(err_lstm, bins=20, alpha=0.5, label=f"LSTM (RMSE={rmse_lstm:.1f})", color="tab:green")

plt.axvline(0, color="black", linestyle="--", linewidth=1)
plt.xlabel("Erreur de prédiction (RUL prédit − RUL réel)")
plt.ylabel("Nombre de moteurs")
plt.title("Distribution des erreurs par modèle\n(à droite de 0 = sur-estimation dangereuse, à gauche = sous-estimation)")
plt.legend()
plt.tight_layout()
plt.savefig("phase6_error_distribution.png", dpi=100)
plt.show()

# Résumé chiffré : part des erreurs de sur-estimation vs sous-estimation
for name, err in [("Random Forest", err_rf), ("LSTM", err_lstm)] + (
    [("XGBoost", err_xgb)] if "err_xgb" in dir() else []):
    over = (err > 0).mean() * 100
    under = (err <= 0).mean() * 100
    print(f"{name:15s} — sur-estimation: {over:.1f}%   sous-estimation: {under:.1f}%   "
          f"erreur moyenne: {err.mean():+.2f} cycles")
```
